6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate)

```python
peopleCount = data[data["salary"] == ">50K"].shape[0]
print("peopleCount count:",peopleCount)

highEducationList = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]
peopleHighEducationCount = data[(data["salary"] == ">50K") & (data["education"].isin(highEducationList))].shape[0]
print("peopleHighEducationCount count:",peopleHighEducationCount)

if(peopleCount == peopleHighEducationCount):
  print("Yes")
else:
  print("No")
```

8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
marriedCount = data[(data["salary"] == ">50K") & (data["marital-status"].str.contains('Married'))].shape[0]
print("marriedCount: ",marriedCount)

unMarriedCount = data[(data["salary"] == ">50K") & (~data["marital-status"].str.contains('Married'))].shape[0]
print("unMarriedCount: ",unMarriedCount)

if(marriedCount > unMarriedCount):
  print("Married")
else:
  print("Not Married")
```

9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?

```python
maxHours = data["hours-per-week"].max()
print("maxHours: ",maxHours)

maxHoursPeople = data[data["hours-per-week"] == maxHours]
maxHoursPeopleCount = maxHoursPeople.shape[0]
print("maxHoursPeopleCount: ",maxHoursPeopleCount)

richMaxHoursPeopleCount = maxHoursPeople[maxHoursPeople["salary"] == ">50K"].shape[0]
maxHoursMaxSalaryPeoplePercent = (richMaxHoursPeopleCount/maxHoursPeopleCount) * 100
print("maxHoursMaxSalaryPeoplePercent: ",maxHoursMaxSalaryPeoplePercent, "%")
```
